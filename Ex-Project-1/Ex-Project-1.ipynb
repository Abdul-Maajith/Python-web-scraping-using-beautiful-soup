# Example Project - Scraping a dummy website!

```python
import requests
from bs4 import BeautifulSoup as bs 
```

```python
url = "https://keithgalli.github.io/web-scraping/"

# Load the webpage content.
r = requests.get(url + "webpage.html")

# Converting to a beautiful soup object
webpage = bs(r.content)
print(webpage)
```

# Grabing all of the social links from the webpages!
* We must Do that in atleast three ways

```python
# 1st way:
links = webpage.select("ul.socials a")
actual_links = [link["href"] for link in links]
print(actual_links)
```

```python
# 2nd way:
ulinks = webpage.find("ul", attrs={"class": "socials"})
links = ulinks.find_all("a")
actual_links = [link["href"] for link in links]
print(actual_links)
```

```python
# 3rd way:
links = webpage.select("li.social a")
actual_links = [link["href"] for link in links]
print(actual_links)
```

# Scraping the table in that website

```python
import pandas as pd

table = webpage.select("table.hockey-stats")[0]
columns = table.find("thead").find_all("th")
column_names = [c.string for c in columns]

table_rows = table.find("tbody").find_all("tr")
l = []
for tr in table_rows:
    td = tr.find_all("td")
    row = [str(tr.get_text()).strip() for tr in td]
    l.append(row)
    
df = pd.DataFrame(l, columns=column_names)
df.loc[df["Team"] != "Did not play"]
```

# Grabbing all the fun facts that use the word "is"

```python
import re

facts = webpage.select("ul.fun-facts li")
facts_with_is = [fact.find(string=re.compile("is")) for fact in facts]
facts_with_is = [fact.find_parent().get_text() for fact in facts_with_is if fact]
facts_with_is
```

# Downloading an image by scraping!

```python
img = webpage.select("div.row div.column img")
img_url = img[0]["src"]
full_url = url + img_url

# Downloading the image data!
img_data = requests.get(full_url).content
with open("lake_como.jpg", "wb") as handler:
    handler.write(img_data)
```

# Solving the mystery challenge!

```python
files = webpage.select("div.block a")
href = [f["href"] for f in files]

for f in href:
    full_url = url + f
    page = requests.get(full_url)
    bs_page = bs(page.content)
    secret_word_element = bs_page.find("p", attrs={"id": "secret-word"})
    print(secret_word_element.string)
    
```
